```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
car_list_response = requests.get("https://www.car.gr/classifieds/cars/?fs=1&condition=used&category=10&category=11&category=12&category=15&category=13&category=18&registration-from=%3E2010")
used_cars_page = car_list_response.text
```

```python
soup = BeautifulSoup(used_cars_page, "html.parser")
row_anchor = soup.find(name="a", class_="row-anchor")
car_link = "https://www.car.gr" + row_anchor.get("href")
car_link
```

```python
car_data_response = requests.get(car_link)
car_data = car_data_response.text
car_data
```

```python
soup2 = BeautifulSoup(car_data, "html.parser")
model_links = soup2.find_all(
    name="span",
    class_="",
)
print(model_links)
```

```python
model_links_text = [item.get_text() for item in model_links]
```
